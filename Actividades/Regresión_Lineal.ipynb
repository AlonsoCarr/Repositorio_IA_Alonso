```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error


# Generar datos de ejemplo
np.random.seed(0)
horas_estudio = np.random.rand(100, 1) * 10
#calificaciones = 2 * horas_estudio + np.random.randn(100, 1) * 2
calificaciones = 2 * horas_estudio + np.random.randint(100, size =1) * 2

# Dividir los datos en conjuntos de entrenamiento y prueba
train_data, test_data, train_labels, test_labels = train_test_split(horas_estudio, calificaciones, test_size= 0.2, random_state= 50)

# Crear y entrenar el modelo de regresión lineal
model = LinearRegression()
model.fit(train_data, train_labels)

# Realizar predicciones con el conjunto de datos de prueba
predicciones = model.predict(test_data)

# Calcular el error cuadrático medio
mse = mean_squared_error(test_labels, predicciones)
print("Error cuadrático medio:", mse)

# Graficar el modelo de regresión lineal y las predicciones
plt.scatter(test_data, test_labels, color='blue', label='Datos de prueba')
plt.plot(test_data, predicciones, color='orange', label='Modelo de regresión lineal')
plt.xlabel('Horas de estudio')
plt.ylabel('Calificación')
plt.title('Modelo de regresión lineal para la predicción de calificaciones')
plt.legend()
plt.show()
```
