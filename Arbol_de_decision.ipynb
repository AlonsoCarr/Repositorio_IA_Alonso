```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn import tree
```

```python
BC = load_breast_cancer()
X = BC.data
y = BC.target
```

```python
BC.target_names
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(X_train, y_train)
```

```python
# Exactitud del modelo en el entrenamiento.
train_accuracy = clf.score(X_train, y_train)
print(train_accuracy)
```

```python
y_pred = clf.predict(X_test)
```

```python
#Predicción 1
print(y_pred[1])
print(y_test[1])
```

```python
#Predicción 2
print(y_pred[20])
print(y_test[20])
```

```python
#Predicción 3
print(y_pred[-1])
print(y_test[-1])
```

```python
#Predicción 4
print(y_pred[6])
print(y_test[6])
```

```python
#Predicción 5
print(y_pred[79])
print(y_test[79])
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:\n", classification_report(y_test, y_pred, target_names=BC.target_names))
```

```python
plt.figure(figsize=(20, 15))
tree.plot_tree(clf, filled=True, feature_names=BC.feature_names, class_names=BC.target_names)
plt.title("Árbol de decisión para el tipo de cancer de mama")
plt.savefig("decision_tree.png", dpi=300)
plt.show()
```
